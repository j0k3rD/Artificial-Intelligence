>Ejercicio: Crear un conjunto de datos para regresion con `from sklearn.datasets import make_regression` con una sola feature, y si quieren le agregan noise (5)

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(
    n_samples=200,
    n_features=1,
    n_targets=1,
    bias=0.8,
    effective_rank=None,
    tail_strength=0.5,
    noise=5,
    coef=False,
    shuffle=False,
    random_state=4231
)

plt.figure(figsize=(7,7))

plt.scatter(
    X[:, 0],
    y,
    color="green",
    edgecolors="k",
    s=100,
    alpha=0.9,
)
```
